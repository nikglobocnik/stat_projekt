```python
import pandas as pd
import numpy as np
import astropy as ast
import matplotlib.pyplot as plt
import statistics as stat
from scipy import stats
from scipy.stats import norm, t
from scipy.stats import gamma
from scipy.special import ndtri

Hali = pd.read_csv("Delnice_Haliburton.csv")
McDon= pd.read_csv("Delnice_McDonalds.csv")
```

Histogram donosov

```python
donosiH = Hali.DONOS
donosiM = McDon.DONOS
```

```python
plt.hist(donosiH, bins = len(donosiH))
```

```python
plt.hist(donosiM, bins = len(donosiH))
```

```python
IQRH = stats.iqr(donosiH, rng=(25, 75), scale=1, nan_policy="omit")
nH = len(donosiH)
sirinaH = 2.6*IQRH/np.cbrt(nH)
strazredovH = int(np.ceil((donosiH.max() - donosiH.min()) / sirinaH))
```

```python
povprecjeH = donosiH.mean()
stdH = donosiH.std()
```

```python
zacetekH =  (int(donosiH.min())//sirinaH)*sirinaH
konecH = (int(donosiH.max())//sirinaH+1)*sirinaH
```

```python
x = np.linspace(donosiH.min(), donosiH.max(), nH)
plt.hist(donosiH, bins=strazredovH, range=(donosiH.min(), donosiH.max()), density=True)
plt.plot(x, stats.norm.pdf(x, povprecjeH, stdH),color='orange')
```

```python
IQRM = stats.iqr(donosiM, rng=(25, 75), scale=1, nan_policy="omit")
nM = len(donosiM)
sirinaM = 2.6*IQRM/np.cbrt(nM)
strazredovM = int(np.ceil((donosiM.max() - donosiM.min()) / sirinaM))
```

```python
povprecjeM = donosiH.mean()
stdM = donosiH.std()
```

```python
zacetekM =  (int(donosiM.min())//sirinaM)*sirinaM
konecM = (int(donosiM.max())//sirinaM+1)*sirinaM
```

```python
x = np.linspace(donosiM.min(), donosiM.max(), nM)
plt.hist(donosiM, bins=strazredovM, range=(donosiM.min(), donosiM.max()), density=True)
plt.plot(x, stats.norm.pdf(x, povprecjeM, stdM),color = 'orange')
```

```python
xH = np.sort(donosiH)
normalnaH = np.random.normal(povprecjeH,stdH,len(xH))
normalnaH.sort()
plt.scatter(xH,normalnaH)
plt.plot([min(xH),max(xH)],[min(xH),max(xH)],color='orange')
```
